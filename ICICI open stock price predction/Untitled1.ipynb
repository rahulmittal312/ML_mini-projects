```python
my_data=quandl.get("NSE/ICICIBANK")
```

```python
my_new=my_data[3200:]
```

```python
train_data=my_new[:-30]
```

```python
train_data.shape
```

```python
test_data=my_new[1691:]
```

```python
test_data=test_data.iloc[:,0:1].values
```

```python
test_data
```

```python
train_data
```

```python
train_data=train_data.iloc[:,0:1].values
```

```python
train_data.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc=MinMaxScaler()
```

```python
train_data=sc.fit_transform(train_data)
```

```python
training_data=train_data[0:1689]
```

```python
train_y=train_data[1:1690]
```

```python
training_data.shape
```

```python
training_data=np.reshape(training_data,(1689,1,1))
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
from keras.layers import LSTM 
```

```python
regressor_2=Sequential()
```

```python
regressor_2.add(LSTM(units=4,activation='sigmoid',input_shape=(None,1)))
```

```python
regressor_2.add(Dense(units=1))
```

```python
regressor_2.compile(optimizer='adam',loss='mean_squared_error')
```

```python
regressor_2.fit(training_data,train_y,batch_size=32,epochs=100)
```

```python
inputs=test_data
```

```python
inputs=sc.transform(inputs)
```

```python
test_data.shape
```

```python
inputs=np.reshape(inputs,(52,1,1))
```

```python
predict_price=regressor_2.predict(inputs)
```

```python
predict_price=sc.inverse_transform(predict_price)
```

```python
predict_price
```

```python
plt.figure(figsize=(12,7))
plt.plot(test_data,color='red',label='Real price')
plt.plot(predict_price,color='blue',label='Predict price')
plt.title('ICICI')
plt.xlabel('Time')
plt.ylabel('Stock price')
plt.legend()
```
